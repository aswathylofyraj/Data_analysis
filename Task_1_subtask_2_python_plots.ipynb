```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file ="/content/data.csv"
data=pd.read_csv(file)
data.head()
```

Income of Individuals of Different Educations

```python
education_income_distribution = data.groupby(['education', 'income']).size().unstack().fillna(0)
```

```python
education_income_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Income Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Individuals')
plt.show()
```

Income of Individuals of Different Working classes

```python
workclass_income_distribution = data.groupby(['workclass', 'income']).size().unstack().fillna(0)
```

```python
workclass_income_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Income Distribution by Workclass')
plt.xlabel('Workclass')
plt.ylabel('Number of Individuals')
plt.show()
```

Income of individuals of Different Occupation with sex category

```python
income_distribution = data.groupby(['sex', 'income']).size().unstack(fill_value=0)
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(income_distribution))

plt.bar(index, income_distribution.loc['Male'], bar_width, label='Male', color='blue', alpha=0.7)

plt.bar([i + bar_width for i in index], income_distribution.loc['Female'], bar_width, label='Female', color='lightcoral', alpha=0.7)


plt.title('Income Distribution by Sex', fontsize=16)
plt.xlabel('Income Category', fontsize=14)
plt.ylabel('Number of Individuals', fontsize=14)
plt.show()
```
